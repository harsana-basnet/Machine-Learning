```python
# Stackings on numpy array.
```

```python
# With hstack we can appened data horizontally.

a = np.array([1,2,3])
b = np.array([4,5,6])

print("a:",a,"\n")
print("b:",b,"\n")
print('Horizontal Append:', np.hstack((a, b)))
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("a:", a, "\n")
print("b:", b, "\n")

print("Vertical Stack:")
print(np.vstack((a, b)))
```

```python
#Basic Statistics on numpy array
```

```python
normal_array = np.random.randn(3, 3)

# Min
print("Min:", np.min(normal_array))

# Max
print("Max:", np.max(normal_array))

# Mean
print("Mean:", np.mean(normal_array))

# Median
print("Median:", np.median(normal_array))

# Standard Deviation
print("Std:", np.std(normal_array))
```

```python
#Matrix Operations
```

```python
# The determinant of an array
import scipy.linalg as slin

a = np.array([[ 3, -5, 8],
 [-1, 2, 3],
 [-5, -6, 2]])

print(slin.det(a))
```

```python
#The trace of an array is found by using the trace() function from numpy.

a = np.array([[ 3, -5, 8],
             [-1, 2, 3],
            [-5, -6, 2]])

print(np.trace(a))
```

```python
# Inverse of a matrix is computed from scipy.linalg.inv() function

a = np.array([[ 3, -5, 8],
                [-1, 2, 3],
               [-5, -6, 2]])
slin.inv(a)
```

```python

a = np.array([[ 3, -5, 8],
           [-1, 2, 3],
           [-5, -6, 2]])
np.transpose(a)
```
