```python
#Ordinal encoding..
from sklearn.preprocessing import OrdinalEncoder
education_order = ['High School','Bachelor','Master','PhD']		
service_order = ['Poor', 'Fair', 'Good', 'Excellent'] 
oe = OrdinalEncoder(categories=[education_order,service_order])
df[['Education_Level','Service_Rating']] = oe.fit_transform(df[['Education_Level','Service_Rating']])

print("After Ordinal Encoding:")
print("Subscription mapping: High School=0,  Bachelor=1  ,Master=2  ,PhD=3")
print("\nRating mapping:      Poor=0, Fair=1, Good=2, Excellent=3")
print()
df
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False)

# Fit and transform City and Payment_Method
encoded_data = ohe.fit_transform(df[['City', 'Payment_Method']])

# Get names of new columns
encoded_columns = ohe.get_feature_names_out(['City', 'Payment_Method'])

print("New columns created:")
print(encoded_columns)

# Convert to DataFrame
encoded_df = pd.DataFrame(
    encoded_data.astype(int),
    columns=encoded_columns,
    index=df.index
)

print("\nOne-Hot Encoded Data:")
print(encoded_df)
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# Load your dataset (IMPORTANT)
df = pd.read_csv("encoding_practice.csv")   # change file name if needed

# Ordinal category order
education_order = ['High School', 'Bachelor', 'Master', 'PhD']
rating_order = ['Poor', 'Fair', 'Good', 'Excellent']

# Column Transformer
ct = ColumnTransformer(
    transformers=[
        (
            'ordinal',
            OrdinalEncoder(categories=[education_order, rating_order]),
            ['Education_Level', 'Service_Rating']
        ),
        (
            'onehot',
            OneHotEncoder(drop='first', sparse_output=False),
            ['City', 'Payment_Method']
        )
    ],
    remainder='passthrough'
)

# Fit and transform data
encoded_data = ct.fit_transform(df)

# Get One-Hot column names
onehot_cols = ct.named_transformers_['onehot'].get_feature_names_out(
    ['City', 'Payment_Method']
)

# Final column names
all_columns = (
    ['Education_Level', 'Service_Rating']
    + list(onehot_cols)
    + ['Customer_ID', 'Purchase_Decision']
)

# Create final DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=all_columns)

print(encoded_df.head())
```
