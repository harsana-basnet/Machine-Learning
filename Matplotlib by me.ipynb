```python
#MATPLOTLIB

# Matplotlib is a Python library for creating visualizations such as line plots, scatter plots, bar plots, histograms, and more. It provides a flexible and customizable interface to help you effectively visualize your data.
```

```python
#Plotting a simple line in matplotlib

import matplotlib.pyplot as plt # or from matplotlib import pyplot as plt
import numpy as np

x=np.arange(0,10,2) #[0, 2, 4, 6, 8]
y=x**2

#making a line by plotting x and y
plt.plot(x,y)

#for setting title of graph
plt.title('graph')

#for x-label
plt.xlabel('x-axis')

#for y-label
plt.ylabel('y-axis')

#for showing the plot
plt.show()
```

```python
#  # MARKER (style, color, size)


# Representation of point in Graph/Chart is Marker

# Types of Marker styles
# ' o ' = circle
# ' s ' = square
# ' d ' = diamond
# ' p ' = pentagon
# ' h ' = hexagon
# ' . ' = point
# ' * ' = star
# ' + ' = plus
# ' | ' = vertical line
# ' _ ' = horizontal line
# ' ^ ' = triangle_up marker
# ' < ' = triangle_left marker
# ' > ' = triangle_right marker
# If we use capital letter it becomes more bold like ( Markers = 'D' instead of Marker='d' )


# For the Colour of the Markers
# We can use single letter predefined colours or use HEX ('#008000'), RGB or RGBA

# The available pre-defined colours are:

# r = red
# b = blue
# w = white
# k = black
# c = cyan
# m = magenta
# y = yellow
# To use the marker and its colour and size we pass in plot() function as parameters:

# plt.plot(x, y, marker="value", mec="value", ms=value)

# mec: marker edge colour
# ms: marker size
```

```python
x=np.arange(0,10,2)
y=x**2

plt.plot(x,y,marker="+",mec="b",ms=25)
plt.show()

# mec= marker edge colour
# ms= marker size
```

```python
# Lines (style, color, width):

# line styles (ls): solid, dotted, dashed, dashdot, none
# line width (lw): lw="value"
# line color (c): same colour as marker colors
# Syntax: plt.plot(x, y) # gives default solid line plt.plot(x, y, ls="value", lw=value, c="value")
```

```python
x=np.arange(0,10,2)
y=x**2

plt.plot(x,y,ls="dashed",lw=7,c='r')
plt.show()
```

```python
# Using the markers and line properties at the same time
```

```python
x=np.arange(0,10,2)
y=x**3

plt.plot(x,y,marker='o',mec='k',ms=20,ls="dashed",lw=7,c='r')
plt.show()
```

```python
 # Font properties for titles and labels

# (font, color, size, position(location), padding)

# If we want to set the font, color, width and size to title and label we give value in fontdict as dictionary: fontdict={'font':'Arial','color':'r', 'size':20}

# Position (location):

# For title and xlabel: left, center and right as values in loc
# For ylabel: top, center and bottom as values in loc
# Padding:

# For title: pad
# For x and y labels: labelpad
```

```python
x=np.arange(0,10,2)
y=x**2

#making a line by plotting x and y
plt.plot(x,y)

#to give value for fontdict
d={'font':'Arial','color':'g','size':16,'weight':'bold'}
d1={'font':'serif','color':'b','size':16}

#for setting title of graph
plt.title('graph',fontdict=d,loc='center',pad=10)

#for x-label
plt.xlabel('x-axis',fontdict=d1,loc='left',labelpad=20)

#for y-label
plt.ylabel('y-axis',fontdict=d1,loc='top',labelpad=60)

#for showing the plot
plt.show()
```

```python
# Grid Function grid()

# It generates the grid lines in the graph.

# For vertical lines through x axis: axis='x'
# For horizontal lines through y axis: axis='y'
# For both: axis='both'
```

```python
# For vertical lines through x axis

x=np.arange(0,10,2)
y=x**2

plt.plot(x,y,marker="+",mec="r",mfc="b",ms=25)
plt.grid(axis='x')
plt.show()
```

```python
# For horizontal lines through y axis

x=np.arange(0,10,2)
y=x**2

plt.plot(x,y,marker="+",mec="r",mfc="b",ms=25)
plt.grid(axis='y')
plt.show()
```

```python
# For both axis

x=np.arange(0,10,2)
y=x**2

plt.plot(x,y,marker="+",mec="r",mfc="b",ms=25)
plt.grid(axis='both')
plt.show()
```

```python
#  SUBPLOTS
# We can make many subplots in a single figure. To make subplots we can use subplot(row, column, index).

# Suppose we want two subplots in a single row, we use row=1, column=2.
# Suppose we want three subplots in a single column, we use row=3, column=1.
# Notes:

# The number of plots is given by multiplying row * column.
# We use suptitle() to give a title to the main heading.
# plt.tight_layout() is used to avoid overlapping between figures.
```

```python
# Making two subplots (1st method)

from matplotlib import pyplot as plt
x1=[1,2,3,4,5]
y1=[1,2,3,4,5]

x2=[5,6,7,8,9,10]
y2=[5,6,7,8,9,10]

plt.subplot(1,2,1) # row=1, col=2, index=1
plt.plot(x1,y1)
plt.title("1st subplot")
plt.xlabel("1x-axis")
plt.ylabel("1y-axis")

plt.subplot(1,2,2) # row=1, col=2, index=2
plt.plot(x2,y2)
plt.title("2nd subplot")
plt.xlabel("2x-axis")
plt.ylabel("2y-axis")

plt.suptitle("figure")
plt.tight_layout() 
plt.show()
```

```python
# Making subplots using plt.subplots()

x1 = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]

x2 = [5, 6, 7, 8, 9, 10]
y2 = [5, 6, 7, 8, 9, 10]

x3 = [5, 6, 7, 8, 9, 10]
y3 = [7, 8, 8, 20, 10, 33]

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))   # Adjust figsize as needed

# Flatten the array of axes objects
axs = axs.flatten()

# Plot on each subplot
axs[0].plot(x1, y1)
axs[0].set_title("Plot 1")
axs[0].set_xlabel("X-axis")
axs[0].set_ylabel("Y-axis")

axs[1].plot(x2, y2)
axs[1].set_title("Plot 2")
axs[1].set_xlabel("X-axis")
axs[1].set_ylabel("Y-axis")

axs[2].plot(x3, y3)
axs[2].set_title("Plot 3")
axs[2].set_xlabel("X-axis")
axs[2].set_ylabel("Y-axis")

# Add overall title and adjust layout
plt.suptitle("Figure Title")
plt.tight_layout()

# Display the plots
plt.show()


# If we have value 1 for row or column plt.subplots(1, 3), 
# it returns a NumPy array of 1D, so we have to access ax[0].plot,
# ax[1].plot(x2, y2), ax[2].plot(x3, y3) like this.
```

```python
# Legend

# Legend function parameters:

# label: we pass list in label inside legend function for how many curve description we want according to our curves
# loc: position of legend box ('best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center')
# fontsize: value
# facecolor: gives the background color to box (e.g., 'y')
# edgecolor: gives the color to the edge (e.g., 'k')
# framealpha: value between 0-1 for opacity of the frame
```

```python
import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]

x2=[1, 2, 3, 4, 5]
y2=[1,4,9,16,25]

plt.plot(x1,y1,x2,y2)
plt.legend(title="line desc",labels=['linear graph','quadradic graph'],loc="center",fontsize=15,
           facecolor='b',edgecolor='r',framealpha=0.2)
plt.show()
```

```python
# BAR GRAPH ( bar(), barh() )
```

```python
## Plotting the vertical bar graph with single colo

x=['BOY','GIRL','MAN','WOMEN','OLD']
y=[100,150,300,500,110]

#for plotting bar graph vertical we use bar()
plt.bar(x,y,color='r',width=0.9) #for horizontal bar we use height instead of width
plt.show()
```

```python
##Plotting the horizontal bar graph with multiple color

x=['BOY','GIRL','MAN','WOMEN','OLD']
y=[100,150,300,500,110]

#for plotting bar graph vertical we use bar()
col=['red','yellow','green','blue','black']

plt.barh(x,y,color=col,height=0.4)   # passing list in c and using height for thickness of horizontal bar diagram

plt.show()
```

```python
 # PLOTTING THE STACKED BAR DIAGRAM
```

```python


# Favourite subjects
x = ['Boy', 'Girl']
sci = [30, 20]
math = [10, 40]
soc = [10, 10]

# Stacked bars
plt.bar(x, sci, label='Science')
plt.bar(x, math, bottom=sci, label='Math')
plt.bar(x, soc, bottom=np.add(sci, math), label='Social')

# Labels and title
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.title("Favourite Subjects by Gender")

# Legend
plt.legend()

# Show plot
plt.show()
```

```python
#  PLOTTING MULTIPLE BAR DIAGRAM

# Categories on x-axis
genres = ['Comedy', 'Action', 'Romance', 'Drama', 'SciFi']

# Number of boys and girls who like each genre
boys = [30, 20, 40, 20, 15]
girls = [20, 40, 20, 15, 25]

# Positions: [0, 1, 2, 3, 4]
x = np.arange(len(genres))

# Width of each bar
width = 0.4

# Boys bars
plt.bar(x, boys, width, label='Boys')

# Girls bars (shifted right by width)
plt.bar(x + width, girls, width, label='Girls')

# Labels and title
plt.xlabel("Genres")
plt.ylabel("Number of People")
plt.title("Favourite Type of Movie")

# Put genres
plt.xticks(x , genres)

# Show legend
plt.legend()

# Display chart
plt.show()
```

```python
# PLOTTING A PIE CHART

student_values=[15,25,12,8]
plt.pie(student_values)
plt.show()
```

```python
# Parameters:

# labels=[]: information of portion of pie
# startangle: degree to begin (moves counter-clockwise)
# explode=[]: takes out portion from pie
# colors=[]: gives colors to the pie
# autopct="%.3f%%": shows percentage with decimal values
```

```python
from matplotlib import pyplot as plt

student_performance=["Excellent","Good","Average","Poor"]
student_values=[15,25,12,8]
c=['r','m','b','y']
plt.figure(figsize=(10, 6))
plt.pie(student_values,labels=student_performance,startangle=36,explode=[0.2,0.3,0,0.1],colors=c,autopct="%.1f%%")
plt.show()
```

```python
# PLOTTING HISTOGRAM

marks=[90,80,80,63,83,23,43,22,55,43,100]
grade_intervals=[0,35,70,100] #(0-35,35-70,70-100)

plt.title("Student Grade")
plt.hist(marks,grade_intervals)
plt.xticks([0,35,70,100],grade_intervals)
plt.xlabel("percentage")
plt.ylabel("No of students")
plt.show()
```
