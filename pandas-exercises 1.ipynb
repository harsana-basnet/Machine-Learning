### 3. Handling Null Values
**Q: Check for null values in the entire DataFrame using and count them per column.**

```python
df.isnull().sum()
```

**Show all rows with an empty value in the name column.**

```python
df[df['name'].isnull()]
```

**Q: Remove rows where the name column is null.**

```python
df = df[~df['name'].isnull()] # df.dropna(subset=[''])
```

**Show all rows with an empty value in the age column.**

```python
df[df['age'].isnull()] #roll - 35,60,65
```

**Q: Impute missing values in the age column with the mean age.**

```python
df['age']=df['age'].fillna(int(df['age'].mean()))
```

**Q: After imputing the missing values, display the rows with roll_no 35, 60, and 65, which previously had missing values.**

```python
df[(df['roll_no']==35)|(df['roll_no']==60)|(df['roll_no']==65)]
```

**Q: See again the count of null values per column**

```python
df.isnull().sum()
```

**Q. Show all rows with an empty value in the gender column.**

```python
df[df['gender'].isnull()]
```

**Q: Fill missing values in the gender column using the mode.**

```python
df['gender'] = df['gender'].fillna(df['gender'].mode()[0]) # roll-> 60
```

**Q: After imputing the missing values, display the rows with 60 which previously had missing gender values.**

```python
df.query('roll_no == 60')
```

**Q: Drop all other null rows permanently.**

```python
df.dropna(inplace=True)
```
